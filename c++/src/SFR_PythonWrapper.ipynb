```python
target = "/Users/thomasvo/eclipse-workspace/SFRcpp/images/GTK_ROI_80d13/"

filenames = np.sort(os.listdir(target))

plt.figure()
for f in filenames:
    if f.endswith('.txt'):
        path = os.path.join(target, f)
        # Opening JSON file
        with open(path) as json_file:
            data = json.load(json_file)

            rows = data['Metadata']['Number of Rows']
            cols = data['Metadata']['Number of Columns']
            byte_stream = np.array(data['Data']['Image Stream'])
            esf = np.array(data['Data']['ESF'])
            mtf = np.array(data['Data']['MTF'])
            freq = np.array(data['Data']['Freq'])
            roi = np.array(data['Data']['Raw Image'])
            
            plt.plot(freq,mtf, label=f.strip('.txt'))

plt.xlim(0,300)
plt.ylim(0,1)
plt.grid()
plt.xlabel('Frequency [lp/mm]')
plt.ylabel('MTF')
```
